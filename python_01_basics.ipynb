```python
#Data Types
# Cell 2 — copy this exactly

# 1. Integer (whole number — no decimal point)
loan_amount = 50000
num_applicants = 614

# 2. Float (number with decimal point)
approval_rate = 68.73
interest_rate = 8.5

# 3. String (text — always inside quotes)
property_area = "Semiurban"
loan_status = "Approved"

# 4. Boolean (only two possible values: True or False)
is_graduate = True
is_self_employed = False

# Print all of them
print(loan_amount)
print(approval_rate)
print(property_area)
print(is_graduate)
```

```python
# To check the data type of each variable

print(type(loan_amount))     # <class 'int'>
print(type(approval_rate))   # <class 'float'>
print(type(property_area))   # <class 'str'>
print(type(is_graduate))     # <class 'bool'>
```

```python
# The Print Statement and f-strings

# Plain print
print(loan_amount)

# f-string — put f before the quote, then {variable_name} inside
print(f"The loan amount is {loan_amount}")
print(f"Approval rate is {approval_rate}%")
print(f"Property area: {property_area}")
print(f"Is graduate: {is_graduate}")

# You can do math inside the curly braces too
print(f"Double the loan: {loan_amount * 2}")
```

```python
# Basic Arithmetic Operations
# Basic math operators

print(50000 + 25000)    # addition → 75000
print(100000 - 30000)   # subtraction → 70000
print(5000 * 12)        # multiplication → 60000
print(75000 / 4)        # division → 18750.0
print(75000 // 4)       # floor division (no decimal) → 18750
print(10 ** 2)          # power (10 squared) → 100
print(round(68.7344, 2))  # round to 2 decimal places → 68.73
```

```python
# List

loan_amounts = [50000, 75000, 30000, 120000, 45000]
property_areas = ["Urban", "Semiurban", "Rural", "Urban", "Semiurban"]
```

```python
# Accessing items by index

loan_amounts = [50000, 75000, 30000, 120000, 45000]

print(loan_amounts[0])   # first item → 50000
print(loan_amounts[1])   # second item → 75000
print(loan_amounts[-1])  # last item → 45000 (negative counts from end)
```

```python
# Useful list functions

loan_amounts = [50000, 75000, 30000, 120000, 45000]

print(len(loan_amounts))   # how many items → 5
print(sum(loan_amounts))   # total → 320000
print(max(loan_amounts))   # largest → 120000
print(min(loan_amounts))   # smallest → 30000

# Calculate average yourself
average = sum(loan_amounts) / len(loan_amounts)
print(f"Average loan amount: {average}")
print(f"Average loan amount (rounded): {round(average, 2)}")
```

```python
# Adding items to a list

loan_amounts = [50000, 75000, 30000]
print(loan_amounts)          # [50000, 75000, 30000]

loan_amounts.append(90000)   # add one item to the end
print(loan_amounts)          # [50000, 75000, 30000, 90000]
```
