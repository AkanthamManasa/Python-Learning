```python
# First function: approval rate calculator

def calculate_approval_rate(approved, total):
    result = (approved / total) * 100
    return result

# Call it with your Bank Loan Project numbers
rate = calculate_approval_rate(422, 614)
print(f"Approval rate: {round(rate, 2)}%")
```

```python
#  Function that returns a category

def categorise_loan(amount):
    if amount > 100000:
        return "Large"
    elif amount > 50000:
        return "Medium"
    else:
        return "Small"

# Test with different values
print(categorise_loan(150000))   # Large
print(categorise_loan(85000))    # Medium
print(categorise_loan(25000))    # Small

# Use it inside an f-string directly
loan = 92000
print(f"₹{loan} is a {categorise_loan(loan)} loan")
```

```python
#  Function with default parameter

def calculate_emi(loan_amount, months=12):
    emi = loan_amount / months
    return round(emi, 2)

# Call with both parameters
print(calculate_emi(120000, 24))   # 5000.0 — 24 monthly payments

# Call with only one — months uses default value of 12
print(calculate_emi(120000))       # 10000.0 — 12 monthly payments
```

```python
# Function + loop combined

def classify_approval_rate(rate):
    if rate >= 70:
        return "High"
    elif rate >= 60:
        return "Medium"
    else:
        return "Low"

areas = ["Urban", "Semiurban", "Rural"]
rates = [70.5, 68.2, 58.7]

for area, rate in zip(areas, rates):
    classification = classify_approval_rate(rate)
    print(f"{area}: {rate}% → {classification} approval zone")
```

```python
# Exercise 1 — Write and call three functions

# Function 1: Check if loan is approved based on credit history and amount
def check_eligibility(credit_history, loan_amount):
    if credit_history == 1 and loan_amount <= 150000:
        return "Eligible"
    elif credit_history == 1 and loan_amount > 150000:
        return "Eligible — needs senior review"
    else:
        return "Not eligible"

# Function 2: Calculate rejection rate
def calculate_rejection_rate(approved, total):
    rejected = total - approved
    rate = (rejected / total) * 100
    return round(rate, 2)

# Function 3: Label credit history
def label_credit(credit_history):
    if credit_history == 1:
        return "Good credit"
    else:
        return "Poor credit"

# Test all three
print(check_eligibility(1, 85000))     # Eligible
print(check_eligibility(1, 200000))    # Eligible — needs senior review
print(check_eligibility(0, 50000))     # Not eligible

print(calculate_rejection_rate(422, 614))   # 31.27

print(label_credit(1))   # Good credit
print(label_credit(0))   # Poor credit
```

```python
# Exercise 3 — Full analysis combining everything

# Dataset
area_data = {
    "Urban":     {"applications": 202, "approved": 142},
    "Semiurban": {"applications": 233, "approved": 179},
    "Rural":     {"applications": 179, "approved": 101}
}

# Function to calculate rate
def get_rate(approved, total):
    return round((approved / total) * 100, 2)

# Function to classify rate
def classify_rate(rate):
    if rate >= 70:
        return "Strong"
    elif rate >= 60:
        return "Average"
    else:
        return "Weak"

# Loop through and analyse
print("AREA-WISE LOAN APPROVAL ANALYSIS")
print("-" * 45)

for area, data in area_data.items():
    rate = get_rate(data["approved"], data["applications"])
    zone = classify_rate(rate)
    print(f"{area:12} | {rate:5}% | {data['approved']:3}/{data['applications']:3} | {zone}")
```
