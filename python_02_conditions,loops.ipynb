```python
# If/else basics

loan_amount = 85000

if loan_amount > 100000:
    print("Large loan")
elif loan_amount > 50000:
    print("Medium loan")
else:
    print("Small loan")
```

```python
# Comparison operators

loan_amount = 85000
credit_history = 1
is_graduate = True

print(loan_amount > 50000)     # Greater than → True
print(loan_amount < 50000)     # Less than → False
print(loan_amount == 85000)    # Equal to → True  (note: TWO equal signs)
print(loan_amount != 100000)   # Not equal → True
print(loan_amount >= 85000)    # Greater than or equal → True
print(loan_amount <= 85000)    # Less than or equal → True
```

```python
#  Combining conditions

loan_amount = 85000
credit_history = 1
property_area = "Urban"

# AND — both conditions must be True
if loan_amount > 50000 and credit_history == 1:
    print("Likely approved — good credit, decent loan size")

# OR — at least one condition must be True
if property_area == "Urban" or property_area == "Semiurban":
    print("Metro area applicant")

# Combining AND + OR
if loan_amount > 50000 and credit_history == 1 and property_area == "Urban":
    print("Strong application")
else:
    print("Needs review")
```

```python
#  Loan categoriser

def categorise_loan(amount):
    if amount > 100000:
        return "Large"
    elif amount > 50000:
        return "Medium"
    else:
        return "Small"

# Test it with different values
print(categorise_loan(150000))   # Large
print(categorise_loan(85000))    # Medium
print(categorise_loan(25000))    # Small
print(categorise_loan(50001))    # Medium  ← boundary check
print(categorise_loan(50000))    # Small   ← exactly 50000 is NOT > 50000
```

```python
#  Basic for loop

loan_amounts = [50000, 75000, 30000, 120000, 45000]

for amount in loan_amounts:
    print(amount)
```

```python
#  Loop + condition combined

loan_amounts = [50000, 75000, 30000, 120000, 45000]

for amount in loan_amounts:
    if amount > 100000:
        print(f"₹{amount} — Large loan")
    elif amount > 50000:
        print(f"₹{amount} — Medium loan")
    else:
        print(f"₹{amount} — Small loan")
```

```python
#  Loop through two lists together

loan_amounts = [50000, 75000, 30000, 120000, 45000]
property_areas = ["Urban", "Semiurban", "Rural", "Urban", "Rural"]

for amount, area in zip(loan_amounts, property_areas):
    print(f"{area}: ₹{amount}")
```

```python
#  Calculate approval rate per area using a loop

areas = ["Urban", "Semiurban", "Rural"]
approval_rates = [70.5, 68.2, 61.3]

print("Area-wise Approval Rates:")
print("-" * 30)

for area, rate in zip(areas, approval_rates):
    if rate >= 70:
        status = "High"
    elif rate >= 65:
        status = "Medium"
    else:
        status = "Low"
    print(f"{area}: {rate}% — {status} approval zone")
```

```python
# Exercise 1 — Credit risk classifier

applicants = [
    ("Ravi", 85000, 1),
    ("Priya", 200000, 0),
    ("Kiran", 40000, 1),
    ("Suma", 150000, 1)
]

# applicants is a list of tuples: (name, loan_amount, credit_history)
# credit_history: 1 = good, 0 = bad

for name, loan, credit in applicants:
    if credit == 1 and loan <= 100000:
        risk = "Low risk"
    elif credit == 1 and loan > 100000:
        risk = "Medium risk"
    else:
        risk = "High risk"
    print(f"{name} | ₹{loan} | Credit: {credit} | {risk}")
```

```python
# Exercise 2 — Find all large loans and count them

loan_amounts = [50000, 75000, 30000, 120000, 45000, 200000, 80000, 35000, 160000]

large_count = 0
small_count = 0
total = 0

for amount in loan_amounts:
    total = total + amount          # running total
    if amount > 100000:
        large_count = large_count + 1
    else:
        small_count = small_count + 1

print(f"Total applications: {len(loan_amounts)}")
print(f"Large loans (>1L): {large_count}")
print(f"Small/Medium loans: {small_count}")
print(f"Total loan value: ₹{total}")
print(f"Average loan: ₹{round(total / len(loan_amounts), 2)}")
```

```python
# Exercise 3 — Recreate this SQL query in Python:
# SELECT property_area, COUNT(*) as count
# FROM loans
# WHERE loan_amount > 50000
# GROUP BY property_area

applications = [
    ("Urban", 85000),
    ("Rural", 30000),
    ("Urban", 120000),
    ("Semiurban", 75000),
    ("Rural", 20000),
    ("Urban", 95000),
    ("Semiurban", 110000)
]

urban_count = 0
rural_count = 0
semiurban_count = 0

for area, amount in applications:
    if amount > 50000:              # WHERE loan_amount > 50000
        if area == "Urban":         # GROUP BY property_area
            urban_count += 1
        elif area == "Rural":
            rural_count += 1
        else:
            semiurban_count += 1

print("Property Area | Count (loans > 50000)")
print("-" * 40)
print(f"Urban         | {urban_count}")
print(f"Rural         | {rural_count}")
print(f"Semiurban     | {semiurban_count}")
```
